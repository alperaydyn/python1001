## Mantıksal Operatörler (Logical Operators)

<div class="alert alert-box alert-info" style="font-size:12pt; padding:15px; line-height:150%;">
    Mantıksal operatörler iki boolean değer arasında karşılaştırma yaparak bu karşılaştırmanın kullanılan operatöre göre doğru veya yanlış olduğu sonucunu döndürürler.
</div>

image.png

```python
x = 8
a = x < 5
b = x > 5
print('a', a, type(a))
print('b', b, type(b))
```

```python
print(a and b)
print(a or b)
print(not(a))
```

### and

Her iki mantıksal ifadenin aynı anda doğru olması durumunda True döner

```python
print(True  and True)
print(True  and False)
print(False and True)
print(False and False)
```

### or

Her iki mantıksal ifadeden en az bir tanesinin doğru olması durumunda True döner

```python
print(True  or True)
print(True  or False)
print(False or True)
print(False or False)
```

### xor (exclusive or)

Her iki mantıksal ifadenin birbirinden farklı olması durumunda True döner. 

Mantıksal ifadelerin karşılaştırıldığı durumlarda != (not equal) ile aynı sonucu verir

```python
import operator
print(operator.xor(True,True))
print(operator.xor(True,False))
print(operator.xor(False,True))
print(operator.xor(False,False))
```

## Kimlik Operatörleri (Identity Operators)

<div class="alert alert-box alert-info" style="font-size:12pt; padding:15px; line-height:150%;">
    Kimlik operatörleri iki değişkenin bellekte aynı yeri gösterip göstermediğinin kontrolünü yapar.
</div>

image.png

```python
x = 500
y = 500
print(id(x))
print(id(y))
x is y
```

```python
x = 256
y = 256
print(id(x))
print(id(y))
print(x is y)
x is y
```

```python
x = 257
y = 257
print(id(x))
print(id(y))
print(x is y)
```

```python
x = [1,2,3]
y = x.copy() # değer olarak tanımlıyorum
print(x, id(x))
print(y, id(y))

print(x == y) # değer kontrolü yapar
print(x is y) # identity (memory adresi) kontrolü yapar
```

## Bitsel Operatörler (Bitwise Operators)

<div class="alert alert-box alert-info" style="font-size:12pt; padding:15px; line-height:150%;">
    Bitsel operatörler, bir tam sayının bitleri üzerinde işlemler yaparlar. Daha çok sistem programlarında ya da düşük seviyeli kodlarda kullanılırlar. Bitsel operatörlerin ortak özellikleri, işleme soktukları tam sayıları bir bütün olarak değil, bit bit (bitwise) ele almalarıdır.
    <a href="https://necatiergin2019.medium.com/bitsel-i%CC%87%C5%9Flemler-bitwise-operations-e1ad30d605cc">Kaynak</a>
</div>

image.png

```python
x = 3        # (0011)
print(f'{x:04b}') # sonucu ikilik sistemde gösterir
y = 5        # (0101)  
print(f'{y:04b}')

z = x & y
print(f'{z:04b}')
```

```python
x = 3        # (0011)
print(f'{x:04b}') # sonucu ikilik sistemde gösterir
y = 5        # (0101)  
print(f'{y:04b}') # sonucu ikilik sistemde gösterir

z = x | y
print(f'{z:04b}') # sonucu ikilik sistemde gösterir
```

```python
x = 3        # (0011)
print(f'{x:04b}') # sonucu ikilik sistemde gösterir
y = 5        # (0101)  
print(f'{y:04b}') # sonucu ikilik sistemde gösterir

z = x ^ y # xor operatörü
print(f'{z:04b}') # sonucu ikilik sistemde gösterir
```

```python
x = 1 
y = ~x
print(f' {x:04b}') 
print(f'{y:04b}') # ikilik sistemdeki bir değerin tersini verir
```

```python
x = 10 
print(f'{x:08b}') 
y = x<<2
print(f'{y:08b}') # her bir bir değerini 2 birim sola kaydırır
```

```python
x = 10 
print(f'{x:08b}') 
y = x>>1
print(f'{y:08b}') # her bir bir değerini 1 birim sağa kaydırır
```
