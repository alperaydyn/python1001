Bir problemin çözümünde izlenecek olan yollar anlamına gelmektedir ve problemin çözümünün düzenli adımlar halinde yazılmasıyla oluşturulur

## Programcı Kimdir?

Bilgisayar programcısı, yazılım kodlarını bilgisayarın izleyebileceği mantıksal bir diziye dönüştüren programlar yazan kişidir. Bunu çeşitli dillerde yazılım programlarını kullanarak yapar.
[kariyer.net](https://www.kariyer.net/pozisyonlar/bilgisayar+programcisi/nedir#:~:text=Bilgisayar%20programc%C4%B1s%C4%B1%2C%20yaz%C4%B1l%C4%B1m%20kodlar%C4%B1n%C4%B1%20bilgisayar%C4%B1n,dillerde%20yaz%C4%B1l%C4%B1m%20programlar%C4%B1n%C4%B1%20kullanarak%20yapar.)

## Programcının Özellikleri

* Analitik düşünme ve organizasyon kabiliyetine sahiptir
* Problemlere karşı yaratıcı çözümler üretir
* Yeni gelişmeleri takip eder, kullandığı programlama dilleri ile ilgili güncel versiyonları takip eder
* Yaptığı günlük işlere programcı gözüyle bakarak çözüm üretmeye çalışır
* Meraklıdır, araştırmayı sever
