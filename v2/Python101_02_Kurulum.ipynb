## Kernel

<div class="alert alert-box alert-white" style="font-size:14pt; padding:15px; line-height:200%">
    <li>İşletim sistemi üzerinde Python'un farklı versiyonları yüklü olabilir</li>
    <li>JupyterHub üzerinde bu farklı versiyonların baz alındığı Kernel'ler oluşturulur. </li>
    <li>Bu kernellere kullanım amaçlarına göre farklı kütüphaneler ve bunların farklı versiyonları yüklenebilir</li>
    <li>Yeni bir notebook açarken ihtiyaç duyulan kernel seçilerek seçilen kernelin bağımsız bir kopyası oluşturulur.</li>
    <li>Her notebook birbirinden farklı kopyalar üzerinde çalışır ve bu kopyaya sadece bu notebook'tan ulaşılabilir.</li>
    <li style="color:#ee3333">Dikkat !: Bir kernel üzerinde yapılan her değişiklik bu kerneli kullanan tüm notebook'larda etkin olacaktır.
</div>

# JupyterHub Kullanımı (!ÖNEMLİ!)

<div class="alert alert-box alert-danger" style="font-size:12pt; padding:50px 50px; line-height:200%">
Ortak bir kullanım alanı olmasından ötürü JupyterHub kullanımında dikkat etmemiz gereken konular var.
    <li> <b>JupyterHub bir veritabanı değildir.</b> <br/>Veritabanından python ortamına veri çekerken veri boyutuna mutlaka dikkat etmemiz gerekir.
    <li> <b>Her açık kalan kernel kaynak tüketir.</b> <br/>İşimiz bittiğinde çalıştığımız notebook'tan <strong>File > Close and Halt</strong> komutu ile hem kerneli hem de notebook'u kapatarak çıkmalıyız.
    <li> Proje özelinde farklı kütüphane versiyonlarına ihtiyaç duyulması durumunda yeni kernel oluştumalı, ilgili ekiplerle paylaşırken bu kernel ile paylaşmalıyız.
    <li> Ortak kullanılan kerneller üzerinde değişiklik yapmamalıyız, her değişiklik ilgili kerneli kullanan tüm notebook'larda etkin olacaktır.
</div>
