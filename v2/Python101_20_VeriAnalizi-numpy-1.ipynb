### numpy dizilerinde veri tipleri

Numpy'ın veri işlemedeki hızının altında yatan bir diğer sebep de kendine özgü veri tiplerini barındırmasıdır. Veri tiplerinin boyutu hafızada kapladığı yer ve buna bağlı olarak da hesaplama hızını etkilemektedir.

* i – integer
* b – boolean
* u – unsigned integer
* f – float
* c – complex float
* m – timedelta
* M – datetime
* O – object
* S – string
* U – unicode string
* V – fixed chunk of memory for other type ( void )

```python
an = np.array([1,2,3])
print(an.dtype)
print(type(an))
```

```python
an = np.array([11,22,3333333], dtype='U')
print(an)
print(an.dtype)
print(id(an))
```

```python
an = np.append(an, ['44444444'])
print(an)
print(id(an))
print(an.dtype)
```

```python
an = an[:-1]
print(an)
print(id(an))
print(an.dtype)
```

### copy, view

```python
def listenin_karesi(b):
    for i in range(len(b)):
        b[i] = b[i]**2
    return b

a = np.array([1, 2, 3, 4])
an = listenin_karesi(a)
print(an)
```

```python
print(a)
```

```python
a = np.array([1, 2, 3, 4])
an = listenin_karesi(a.copy())
print(an)
print(a)
```

```python
a = np.array([1, 2, 3, 4])
an = listenin_karesi(a.view())
print(an)
print(a)
```
